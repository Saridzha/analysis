```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler().fit(df)
df = sc.transform(df)
pd.DataFrame(df, columns = ['Область','Кількість ІТ-фахівців, осіб', 'Середній дохід інженера-програміста середнього рівня, грн', 'Середній дохід по всьому регіону на одну особу, грн', 'Сплачені податки ІТ-галуззю, грн', 'Сплачені податки регіоні, грн', 'Зайняте населення по регіонах, осіб', 'Сплачені податки одного фахівця ІТ-галузі, грн', 'Сплачені податки в регіоні на одну зайняту особу, грн']).round()
```

```python
X=df[:, [2, 3, 7, 8]]
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 6):
    kmeans = KMeans(n_clusters = k).fit(X)  
    preds = kmeans.fit_predict(X)
    sil.append(silhouette_score(df, preds, metric = 'euclidean'))


plt.plot(range(2, 6), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
    print(str(i+2) +":"+ str(sil[i]))
```

```python
km = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0).fit_predict(X)
kms = KMeans(n_clusters = 3, random_state = 0).fit(X)
pd.DataFrame(kms.cluster_centers_, columns = ['Середній дохід інженера-програміста середнього рівня, грн', 'Середній дохід по всьому регіону на одну особу, грн', 'Сплачені податки одного фахівця ІТ-галузі, грн', 'Сплачені податки в регіоні на одну зайняту особу, грн']).round(0)
```

```python
from IPython.display import display, HTML



kms = KMeans(n_clusters=3, random_state=0)


kms.fit(X)


clusters = kms.labels_


data_with_clusters = pd.DataFrame(original_data, columns=['Область','Середній дохід інженера-програміста середнього рівня, грн', 'Середній дохід по всьому регіону на одну особу, грн', 'Сплачені податки одного фахівця ІТ-галузі, грн', 'Сплачені податки в регіоні на одну зайняту особу, грн'])
data_with_clusters['Кластер'] = clusters

for cluster_id in range(kms.n_clusters):
    print(f"Зміст кластера {cluster_id}:")
    cluster_data = data_with_clusters[data_with_clusters['Кластер'] == cluster_id].drop(columns='Кластер')
    display(HTML(cluster_data.to_html(index=False)))
    print("\n")
```

```python
кількість_країн_у_кожному_кластері = pd.Series(clusters).value_counts().reset_index()
кількість_країн_у_кожному_кластері.columns = ['Кластер', 'Кількість країн']
кількість_країн_у_кожному_кластері = кількість_країн_у_кожному_кластері.sort_values(by='Кластер')

display(HTML(кількість_країн_у_кожному_кластері.to_html(index=False)))
```

```python
cluster_means = data_with_clusters.groupby('Кластер').mean()
display(HTML(cluster_means.to_html()))
```
